```python
farms = pd.read_csv('../data/farm_csv.csv')
farms['fips'] = farms['fips'].astype(str)
```

```python
def len4(x):
    if len(x) == 4:
        return True
    return False

def add0(x):
    if len4(x):
        return '0' + str(x)
    return x

farms['GEOID'] = farms['fips'].apply(lambda x: add0(x))
geo_farms =  counties.merge(farms, on = 'GEOID')
```

```python
# continental us
alaska_fip = '02'
hawiaii_fip = '15'
geo_farms = geo_farms[~geo_farms['STATEFP'].isin([alaska_fip,hawiaii_fip])]
```

```python
gdf = geopandas.GeoDataFrame(
    data, geometry=geopandas.points_from_xy(data.lng_avg, data.lat_avg))
al = 'Alaska'
ha = 'Hawaii'
gdf = gdf[~gdf['state_name'].isin([al,ha])]
```

```python
fig, ax = plt.subplots(figsize=(10,10))
geo_farms.plot(ax=ax,color='white')
gdf.plot(column='p_per_capita', ax=ax,cmap='Reds',legend=True,scheme='quantiles',marker='o',markersize = 800)
# ctx.add_basemap(ax, zoom=10, source=ctx.providers.Stamen.TonerLite)
ax.set_axis_off()
# ax.set_title('Number of Farms by County and City Produce\n per Capita for Continental US Cities', fontsize=40)
plt.savefig('../img/legend')
```

```python
fig, ax = plt.subplots(figsize=(40,40))
geo_farms.plot(column='number_farms',legend=False, ax=ax,cmap = 'Blues',legend_kwds={'shrink': 0.1})
gdf.plot(column='p', ax=ax,cmap='Reds',scheme='quantiles', legend=True,marker='o',markersize = 800)
ax.set_axis_off()
ax.set_title('Number of Farms by County and City Produce\n Total for Continental US Cities', fontsize=40)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(40,20))

def usmap(col,lab,ax):
    geo_farms.plot(column='number_farms',scheme='quantiles',legend=True, ax=ax,cmap = 'Blues')
    gdf.plot(column=col, ax=ax,cmap='Reds',scheme='quantiles', legend=True,marker='o',markersize = 500,alpha=0.9)
    ax.set_axis_off()
    ax.set_title(f'Number of Farms by County and City Produce\n {lab} for Continental US Cities', fontsize=20)

cols = ['p','p_per_capita']
titles = ['Total','Per Capita']
usmap(cols[1],titles[1],ax)
    
plt.savefig('../img/maps2')
```

```python
fig, axes = plt.subplots(2,2, figsize=(20,15))
plts = axes.ravel()

sns.countplot(x='census_region', data=data, ax=plts[0])

plts[0].set_xlabel('Census Region')
plts[0].set_ylabel('Number of Cities')
plts[0].set_title('Number of Cities vs. Census Region')

def box(y,ylab,ax):
    sns.boxplot('census_region',y, data=data,ax=ax)
    ax.set_xlabel('Census Region')
    ax.set_ylabel(ylab)
    ax.set_title(f'{ylab} vs. Census Region')
y = ['pop2020','p_per_capita','kwh_loss_pct']
titles = ['Population','Produce Per Capita','Kilowatts / Hour Trade Off Pct']
i = 0
for p in plts[1:]:
    box(y[i],titles[i],p)
    i+=1
    
plt.savefig('../img/boxplots')
```

```python
bootstraps = 1000
n = data.shape[0]
kwhs = []
p = []
for _ in range(bootstraps):
    # Resample data
    bootstrap = data.iloc[np.random.randint(n, size=n)]
    p.append(bootstrap['p'].sum())
    kwhs.append(bootstrap['kwh_loss'].sum())
```

```python
fig, axes = plt.subplots(1,2,figsize=(20,7))
sns.distplot(p,rug=True,ax=axes[0])
axes[0].set_title('Distribution of Bootstrapped Estimated Combined Seasonal Produce for Top \n200 Largest Cities Only for Flat Roofs >=10,000 sq ft')
axes[0].set_xlabel('Pounds of Produce (Billions)')
axes[0].set_ylabel('Frequency');
sns.distplot(kwhs,rug=True,ax=axes[1],color='Purple')
axes[1].set_title('Distribution of Bootstrapped Estimated Net Annual Electricity Trade-Off for Top \n200 Largest Cities Only for Flat Roofs >=10,000 sq ft')
axes[1].set_xlabel('Electricity in KWHs (Billions)')
axes[1].set_ylabel('Frequency');
plt.savefig('../img/boots')
```
